1. What is a random variable in probability theory?
A random variable is a numerical outcome of a random experiment. It assigns a number to every possible outcome in a sample space.

2. What are the types of random variables?

Discrete random variable → Takes countable values (e.g., number of heads in coin tosses).

Continuous random variable → Takes values from a continuous range (e.g., height, weight).

3. What is the difference between discrete and continuous distributions?

Discrete distribution: Deals with discrete random variables, probability is assigned to exact values (e.g., binomial distribution).

Continuous distribution: Deals with continuous variables, probability is assigned over an interval using a density function (e.g., normal distribution).

4. What are probability distribution functions (PDF)?
A Probability Distribution Function (PDF) describes how probabilities are distributed over the values of a random variable.

For discrete variables: it is the Probability Mass Function (PMF).

For continuous variables: it is the Probability Density Function.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

PDF/PMF gives the probability at an exact point (or density for continuous).

CDF gives the probability that a variable is ≤ a certain value.

6. What is a discrete uniform distribution?
A distribution where all outcomes are equally likely over a finite set. Example: Rolling a fair die (each face has probability 1/6).

7. What are the key properties of a Bernoulli distribution?

Two outcomes: success (1) or failure (0).

Probability of success = p, failure = 1 – p.

Mean = p, Variance = p(1 – p).

8. What is the binomial distribution, and how is it used in probability?
The binomial distribution models the number of successes in n independent Bernoulli trials with probability p.

11. What are the characteristics of a normal distribution?

Symmetrical and bell-shaped curve.

Mean = Median = Mode.

Defined by mean (μ) and standard deviation (σ).

Total probability = 1.

Most data lies within ±1σ, ±2σ, ±3σ (68–95–99.7 rule).

12. What is the standard normal distribution, and why is it important?

A normal distribution with mean = 0 and standard deviation = 1.

Used to standardize data (via Z-scores).

Important because many statistical tests (Z-test, hypothesis testing) are based on it.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the original population distribution.

Critical because it allows us to use normal distribution for inference even when the population isn’t normal.

14. How does the Central Limit Theorem relate to the normal distribution?
The CLT explains why the normal distribution is so common in statistics: averages of large samples tend to follow a normal distribution, even if the underlying data is not normal.

15. What is the application of Z statistics in hypothesis testing?
Z-statistics are used to test hypotheses when:

The population standard deviation (σ) is known.

The sample size is large (n > 30).
It helps determine if a sample mean significantly differs from the population mean.

17. What are point estimates and interval estimates in statistics?

Point estimate: A single value that estimates a population parameter (e.g., sample mean for population mean).

Interval estimate: A range of values (with confidence level) likely to contain the parameter (e.g., 95% confidence interval).

18. What is the significance of confidence intervals in statistical analysis?

They provide a range of values within which the true population parameter is likely to fall.

Reflect the precision of the estimate.

Wider intervals = less precision, narrower intervals = more precision.

21. What are the assumptions for applying the Central Limit Theorem?

Samples must be independent and randomly selected.

Sample size should be sufficiently large (n ≥ 30 is common rule).

Population should have a finite variance.

For smaller samples, the population should not be heavily skewed.

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to calculate and plot the standard normal distribution
def standard_normal_distribution():
    x = np.linspace(-4, 4, 1000)   # Range of values
    y = norm.pdf(x, 0, 1)          # PDF with mean=0, std=1

    plt.plot(x, y, color='blue', linewidth=2, label="Standard Normal PDF")
    plt.fill_between(x, y, alpha=0.2, color="skyblue")  # Shaded area
    plt.title("Standard Normal Distribution (μ=0, σ=1)")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function
standard_normal_distribution()
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        return 0
    return p**x * (1-p)**(1-x)

# Example usage
p = 0.6
print("P(X=1):", bernoulli_pdf(1, p))
print("P(X=0):", bernoulli_pdf(0, p))
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

import numpy as np
import matplotlib.pyplot as plt

n = 6
x = np.arange(1, n+1)
cdf = np.cumsum(np.ones(n)/n)

plt.step(x, cdf, where='post')
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.title("CDF of Discrete Uniform Distribution")
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

data = np.random.normal(50, 10, 100)
z = z_scores(data)

plt.plot(z, marker='o', linestyle='')
plt.axhline(0, color='red')
plt.xlabel("Index")
plt.ylabel("Z-score")
plt.title("Z-scores of Dataset")
plt.show()
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

data = np.random.normal(50, 10, 50)
sample_mean = np.mean(data)
population_mean = 52
std = np.std(data, ddof=1)
n = len(data)

z = (sample_mean - population_mean) / (std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z-statistic:", z)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as st

data = np.random.normal(50, 10, 100)
mean = np.mean(data)
conf_interval = st.norm.interval(0.95, loc=mean, scale=st.sem(data))

print("95% Confidence Interval:", conf_interval)
```

```python
#23. Implement Z-score normalization on a dataset and visualize its transformation

data = np.random.normal(60, 15, 100)
z_data = (data - np.mean(data)) / np.std(data)

plt.subplot(1, 2, 1)
plt.hist(data, bins=20, edgecolor='black')
plt.title("Original Data")

plt.subplot(1, 2, 2)
plt.hist(z_data, bins=20, edgecolor='black')
plt.title("Z-score Normalized Data")

plt.show()
```
